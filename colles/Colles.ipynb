```python
# Test 3/12/2024 utiliser seulement une fonction avec le jeu de data en argument
import pandas as pd
from icalendar import Calendar, Event
from datetime import datetime
import io

def pdc(data):
    """
    In : data sous forme de jeux de données (Total = 4)
    Out : Fichier Ical format ics éditable en txt
    """

    df = pd.read_csv(io.StringIO(data))

    # Créer un objet Calendrier
    cal = Calendar()

    # Définir la matière
    matiere = df.iloc[1,0].strip()

    # Parcourir les lignes du DataFrame à partir de la ligne 3 (index 2)
    for index, ligne in df.iloc[2:].iterrows():
      # Extraire les informations
      professeur = ligne.iloc[0].strip()
      info_date_heure = ligne.iloc[1]
      salle = ligne.iloc[2]

      # Parcourir les colonnes de dates à partir de la 4ème colonne (index 3)
      for i in range(3, len(df.columns)):
        semaine = df.columns[i]
        valeur = ligne.iloc[i]

        # Vérifier si la valeur est celle demandée
        if valeur == group:
          # Vérifier si la cellule contient le format attendu et n'est pas vide
          if isinstance(info_date_heure, str) and " " in info_date_heure:
            jour, horaire = info_date_heure.split(' ')
            heure_debut, heure_fin = horaire.split('-')

            # Calculer la date complète de l'événement
            # Calculer le jour relatif
            if jour == 'Lu':
              jour_semaine = 0
            elif jour == 'Ma':
              jour_semaine = 1
            elif jour == 'Me':
              jour_semaine = 2
            elif jour == 'Je':
              jour_semaine = 3
            elif jour == 'Ve':
              jour_semaine = 4
            # Calculer la date complete
            date_event = datetime.strptime(semaine, '%d/%m/%Y') + pd.DateOffset(days=jour_semaine)

            # Définir l'heure de début et de fin de l'événement
            heure_debut = int(heure_debut)
            heure_fin = int(heure_fin)

            # Créer un objet Événement
            event = Event()
            event.add('summary', f"Colle {matiere} {professeur}")
            event.add('location', salle)
            event.add('dtstart', datetime(date_event.year, date_event.month, date_event.day, heure_debut, 0))
            event.add('dtend', datetime(date_event.year, date_event.month, date_event.day, heure_fin, 0))
            cal.add_component(event)

    return (cal.to_ical())

def savefile(nomfichier):
    """
    In : filename
    Out : Ics file
    """
    assert type(nomfichier) == str
    jeu1 = pdc('''
    ,horaire,salle,2/9/2024,9/9/2024,16/9/2024,23/9/2024,30/9/2024,7/10/2024,14/10/2024,,4/11/2024,11/11/2024,18/11/2024,25/11/2024,2/12/2024,9/12/2024,16/12/2024,,6/1/2025,13/1/2025,20/1/2025,27/1/2025,3/2/2025,10/2/2025,17/2/2025
    ,,,,,,,,,,,,,,,,,,,,,,,,,
    Anglais,,,,,,,,,,,,,,,,,,,,,,,,,
    Mme Tschudi,Lu 12-13,E02 E04,,,,1,2,3,4,,5,6,7,8,9,14,11,,12,13,14,1,2,3,2
    Mme Mc Grath,Lu 12-13,E20,,,,3,4,13,6,,7,8,9,10,11,12,7,,14,7,16,15,6,1,4
    Mme Mc Grath,Lu 13-14,E20,,,,5,12,1,14,,15,16,11,2,13,10,5,,8,1,2,3,4,5,6
    Mme Rébillon,Lu 13-14,E08 E10,,,,7,8,9,8,,9,4,3,14,15,16,13,,2,15,4,5,16,11,8
    M. Coquilhat,Lu 13-14,K08 K09,,,,9,10,5,12,,13,14,15,16,1,2,3,,16,11,6,7,8,9,10
    Mme Heikka,Lu 18-19,E01 E05,,,,15,14,11,16,,11,10,1,12,3,4,15,,6,3,10,9,10,15,12
    Mme Otéro,Ma 17-18,B21 B22,,,,13,6,15,10,,1,12,13,6,5,8,1,,4,9,8,13,12,7,14
    Mme Otéro,Ma 18-19,B21 B22,,,,11,16,7,2,,3,2,5,4,7,6,9,,10,5,12,11,14,13,16
    ''')
    jeu2 = pdc('''
    ,horaire,salle,2/9/2024,9/9/2024,16/9/2024,23/9/2024,30/9/2024,7/10/2024,14/10/2024,,4/11/2024,11/11/2024,18/11/2024,25/11/2024,2/12/2024,9/12/2024,16/12/2024,,6/1/2025,13/1/2025,20/1/2025,27/1/2025,3/2/2025,10/2/2025,17/2/2025
    ,,,,,,,,,,,,,,,,,,,,,,,,,
    Maths,,,,,,,,,,,,,,,,,,,,,,,,,
    M. Sautonie,Lu 12-13,E01,,,,,16,7,2,,11,10,5,6,7,4,9,,10,3,12,9,14,15,16
    M. Falconnet,Lu 13-14,F106,,,,,14,11,16,,1,2,1,12,5,6,15,,6,5,10,11,12,13,14
    Mme Plà,Ma 12-13,E01,,,,,8,1,14,,9,16,7,2,3,12,5,,14,7,8,13,4,3,12
    Mme Plà,Ma 13-14,E01,,,,,10,15,12,,5,14,15,10,1,8,3,,4,9,6,1,8,9,10
    M. Magniez,Ma 13-14,F103,,,,,6,5,10,,3,12,13,8,15,16,1,,2,11,4,5,16,1,2
    M. Sageaux,Ma 17-18,F205,,,,,4,13,8,,7,4,11,4,13,10,7,,16,13,2,15,10,5,4
    M. Raymond,Ma 18-19,E04,,,,,12,9,6,,15,8,9,16,11,14,13,,12,15,16,7,2,11,6
    Mme Deville,Je 17-18,B21,,,,,2,3,4,,13,6,3,14,9,2,11,,8,1,14,3,6,7,8
    ''')
    jeu3 = pdc('''
    ,horaire,salle,2/9/2024,9/9/2024,16/9/2024,23/9/2024,30/9/2024,7/10/2024,14/10/2024,,4/11/2024,11/11/2024,18/11/2024,25/11/2024,2/12/2024,9/12/2024,16/12/2024,,6/1/2025,13/1/2025,20/1/2025,27/1/2025,3/2/2025,10/2/2025,17/2/2025
    ,,,,,,,,,,,,,,,,,,,,,,,,,
    SI,,,,,,,,,,,,,,,,,,,,,,,,,
    Mme Pradier,Lu 12-13,B21,,,,2,11,4,3,,2,5,8,1,10,9,12,,11,8,13,16,15,2,11
    M. Lisle,Lu 12-13,Labo PT,,,,4,3,6,5,,8,7,10,15,8,11,6,,13,16,15,8,1,4,3
    Mme Pradier,Lu 13-14,B21,,,,6,5,2,7,,14,9,12,11,4,13,16,,15,2,1,12,3,6,5
    M. Boyer,Lu 13-14,Labo PT,,,,8,7,10,9,,12,11,4,13,16,7,2,,1,6,9,6,5,8,7
    M. Lisle,Ma 12-13,Labo PT,,,,16,9,12,11,,10,3,16,9,2,15,10,,3,14,5,4,7,10,15
    M. Deluche,Ma 18-19,Labo PT,,,,12,15,14,13,,16,15,6,7,14,3,14,,5,4,7,2,13,16,1
    M. Vallenet,Je 17-18,Labo PT,,,,14,13,16,15,,6,1,14,3,6,5,8,,7,12,11,14,11,14,13
    M. Deluche,Ve 13-14,Labo PT,,,,10,1,8,1,,4,13,2,5,12,1,4,,9,10,3,10,9,12,9
    ''')
    jeu4 = pdc('''
    ,horaire,salle,2/9/2024,9/9/2024,16/9/2024,23/9/2024,30/9/2024,7/10/2024,14/10/2024,,4/11/2024,11/11/2024,18/11/2024,25/11/2024,2/12/2024,9/12/2024,16/12/2024,,6/1/2025,13/1/2025,20/1/2025,27/1/2025,3/2/2025,10/2/2025,17/2/2025
    ,,,,,,,,,,,,,,,,,,,,,,,,,
    Physique,,,,,,,,,,,,,,,,,,,,,,,,,
    Mme Lefort,Lu 12-13,H105,,,,,15,8,1,,10,3,16,9,2,1,10,,9,4,11,14,13,16,15
    M. Raimi,Lu 12-13,H108,,,,,13,16,15,,6,1,14,3,6,5,8,,7,10,3,2,11,14,13
    M. Qadri,Lu 12-13,B23,,,,,9,14,13,,16,15,2,7,14,15,14,,3,14,7,10,9,12,1
    M. Scotto,Ma 17-18,H111,,,,,1,6,11,,4,13,10,5,12,7,4,,5,12,5,16,7,6,5
    Mme Lines,Me 17-18,H105,,,,,11,4,9,,8,11,4,15,4,11,6,,15,6,9,4,15,8,9
    M. Trollet,Me 18-19,H105,,,,,7,2,7,,14,5,12,11,16,13,16,,1,16,15,6,5,10,3
    M Fayolle,Ve 12-13,E102,,,,,3,12,5,,12,7,6,13,8,3,2,,13,2,1,8,1,4,7
    M Fanjeaux,Ve 12-13,H113,,,,,5,10,3,,2,9,8,1,10,9,12,,11,8,13,12,3,2,11
    ''')
    # Enregistrer le calendrier complet dans un fichier .ics (opération de sommation)
    with open(nomfichier, 'ab') as f:
        f.write(jeu1)
        f.write(jeu2)
        f.write(jeu3)
        f.write(jeu4)
    cleaner(nomfichier)

def cleaner(nomfichier):
    """
    In : Fichier Ical Global après sommation
    Out : Fichier Ical fonctionnel et prêt
    """
    # Read the file and filter lines
    with open(nomfichier, 'r') as f:
        lines = f.readlines()

    # Remove unwanted lines
    filtered_lines = [line for line in lines if line.strip() not in ("BEGIN:VCALENDAR", "END:VCALENDAR")]

    # Add BEGIN:VCALENDAR and END:VCALENDAR
    filtered_lines.insert(0, "BEGIN:VCALENDAR\n")  # Add to the beginning
    filtered_lines.append("END:VCALENDAR\n")     # Add to the end

    # Write the modified lines back to the file
    with open(nomfichier, 'w') as f:
        f.writelines(filtered_lines)

def export():
    """
    In : none
    Out : none, envoi les fichiers vers le bon répertoire, et séparément
    """
    global group
    group = 1
    for i in range(1,17):
        group = i
        savefile(f"../Documents/Cours lycée/Archive/Première 2022:2023/NSI/jeanprt.github.io/colles/Colles groupe {round(group)}.ics")
    print("success")
"""
def run(groupe):
    global group
    group = groupe
    savefile(f"Colles/Colles groupe {round(group)}.ics")
    print("success")
run(14)
"""
export()
```
